```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = './data/weather_prediction_dataset.csv'
df = pd.read_csv(file_path)
```

```python
#df.head()
df["DATE"] = pd.to_datetime(df["DATE"], format="%Y%m%d")
```

```python
# Extract Basel-related columns
basel_data = df[['DATE', 'BASEL_temp_mean', 'BASEL_temp_min', 'BASEL_temp_max', 
                 'BASEL_precipitation', 'BASEL_sunshine', 'BASEL_humidity', 'BASEL_pressure']]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(basel_data['DATE'], basel_data['BASEL_temp_mean'], label='Mean Temperature', color='blue')
plt.plot(basel_data['DATE'], basel_data['BASEL_temp_min'], label='Min Temperature', color='green')
plt.plot(basel_data['DATE'], basel_data['BASEL_temp_max'], label='Max Temperature', color='red')
plt.title('Basel Temperature Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(basel_data['DATE'], basel_data['BASEL_precipitation'], label='Precipitation (mm)', color='blue')
plt.plot(basel_data['DATE'], basel_data['BASEL_sunshine'], label='Sunshine (hours)', color='orange')
plt.title('Basel Precipitation and Sunshine Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(basel_data.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between Basel Weather Variables')
plt.show()
```
